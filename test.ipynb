```python
products = {
    'eggs': 15.99,
    'bread': 20,
    'milk': 100,
    'water': 50
}

stocks = {
    'eggs': '10%',
    'bread': '20%',
    'milk': '15%'
}

def apply_discounts(products, stocks):
    discounts = {}
    for key in products: 
        if key in list(stocks.keys()):
            discounts[key] = round(products[key] - products[key] * int(stocks[key][:2]) / 100, 2)
    else:
        discounts[key] = products[key]
    return discounts

apply_discounts(products, stocks)
```

```python
import pandas as pd
import numpy as np
solar_data = pd.read_csv('solarpower.csv')
solar_data.head()
```

```python
for i in range(len(solar_data['cum_power'])):
    if solar_data['cum_power'][i] == -1:
        solar_data['cum_power'][i] = (solar_data['cum_power'][i-1] + solar_data['cum_power'][i+1]) / 2 

solar_data['cum_power'].mean()       
```

```python
for i in range(1, len(solar_data['cum_power'])):
    solar_data['day_power'][i] = solar_data['cum_power'][i] - solar_data['cum_power'][i-1]
solar_data['day_power'][0] = 0
solar_data
```

```python
solar_data['date'] = pd.to_datetime(solar_data['date'], dayfirst=True)
solar_data["year_month"] = solar_data["date"].dt.to_period("M")
solar_data
```
